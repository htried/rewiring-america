```python
import pandas as pd
```

```python
 # from http://www2.census.gov/programs-surveys/ahs/2019/AHS%202019%20National%20PUF%20v1.1%20CSV.zip
df = pd.read_csv('./household.csv')
```

```python
# from https://www.census.gov/data-tools/demo/codebook/ahs/ahsdict.html
# query parameters:
#   survey year: 2019 national
#   topics/subtopics: geography, demographics, equipment and appliances, utilities, income, 
#   availability: PUF
relevant_col_df = pd.read_csv('./ahsdict_25FEB21_12_18_01_80.csv')
```

```python
shared_cols = []
for var in relevant_col_df['Variable'].tolist():
    if var in df.columns:
        shared_cols.append(var)

shared_ahs = df[shared_cols]
```

```python
for col in shared_ahs.columns:
    try:
        shared_ahs[col] = shared_ahs[col].str.split("'").str[1].tolist()
    except:
        continue
    shared_ahs[col].astype(str)
```

```python
shared_ahs
```

```python
mapped_ahs = shared_ahs
for col in shared_ahs.columns:
    row = relevant_col_df.loc[relevant_col_df['Variable'] == col]
    desc = row['Description'].values[0]
    codes = row['Response Codes'].values[0].split('||')
    map_ = {}
    for code in codes:
        [k, v] = code.split(':', 1)
        if 'or' in k:
            [k1, k2] = k.split(' or ')
            k1 = k1.strip()
            k2 = k2.strip()
            map_[k1] = v
            map_[k2] = v
            continue
        k = k.strip()
        map_[k] = v
    mapped_ahs[col] = shared_ahs[col].map(map_).fillna(shared_ahs[col])
```

```python
mapped_ahs['HINCP'] = mapped_ahs['HINCP'].astype(int)
mapped_ahs['FINCP'] = mapped_ahs['FINCP'].astype(int)
mapped_ahs
```

```python
# Now that data is clean, start analysis.
income_bins = [0, 20000, 40000, 60000, 80000, 100000, 120000, 140000, float('inf')]
regions = ["National",
           " New England",
           " Middle Atlantic",
           " East North Central",
           " West North Central",
           " South Atlantic",
           " East South Central",
           " West South Central",
           " Mountain",
           " Pacific"]
modes_of_analysis = ['HEATTYPE',
                     'HEATFUEL',
                     'HOTWATER',
                     'COOKFUEL',
                     'DRYER']
```

```python
# get percentage breakdowns for a type of fuel source
def percent_df(df):
    percent_dict = {}
    for i in range(1, 150000, 20000):
        percent_list = []
        for row in df[i]:
            percent_list.append(row / df[i].sum())
        if i < 140000:
            percent_dict[f'income ${i-1}-{i+19999}'] = percent_list
        else:
            percent_dict[f'income ${i-1}+'] = percent_list
    return pd.DataFrame(percent_dict, index=df.index)
```

```python
# Step 1.1: How many households are there in each income bracket?
mapped_ahs.HINCP.groupby(pd.cut(mapped_ahs.HINCP, income_bins)).size()
```

```python
# Step 1.2: How many households are there total?
mapped_ahs.HINCP.groupby(pd.cut(mapped_ahs.HINCP, income_bins)).size().sum()
```

```python
def conduct_analysis():
    analysis_dict = {}
    
    for region in regions:
        analysis_dict[region] = {}
        for mode in modes_of_analysis:
            analysis_df = pd.DataFrame()
            if region != "National":
                analysis_df = mapped_ahs.loc[mapped_ahs.DIVISION == region][['HINCP', mode]]
            else:
                analysis_df = mapped_ahs[['HINCP', mode]]
            analysis_groups = analysis_df.groupby([mode, pd.cut(analysis_df.HINCP, income_bins)]).size().unstack()
            analysis_groups_percent = percent_df(analysis_groups)
            total_by_income = analysis_groups.sum(axis=1)
            
            analysis_dict[region][mode] = [analysis_groups, analysis_groups_percent, total_by_income]
    return analysis_dict

analysis_dict = conduct_analysis()
```

```python
# I've abstracted all of the logic from my analysis into a function and applied it programmatically. 
# Here's how to use this dict:
#   analysis_dict[<region>][<mode>][<0, 1, 2>]
#   regions: see the region list above. leading space is necessary as I haven't been able to get it to stop showing up.
#   mode: see mode list above.
#   <0, 1, 2>:
#        0 — raw counts of each category within each income bracket
#        1 — percentages of each category within each income bracket
#        2 — raw count of each category without income analysis
```
