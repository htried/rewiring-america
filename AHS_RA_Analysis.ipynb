```python
import pandas as pd
```

```python
 # from http://www2.census.gov/programs-surveys/ahs/2019/AHS%202019%20National%20PUF%20v1.1%20CSV.zip
df = pd.read_csv('./household.csv')
```

```python
# from https://www.census.gov/data-tools/demo/codebook/ahs/ahsdict.html
# query parameters:
#   survey year: 2019 national
#   topics/subtopics: geography, demographics, equipment and appliances, utilities, income, 
#   availability: PUF
relevant_col_df = pd.read_csv('./ahsdict_25FEB21_12_18_01_80.csv')
```

```python
shared_cols = []
for var in relevant_col_df['Variable'].tolist():
    if var in df.columns:
        shared_cols.append(var)

shared_ahs = df[shared_cols]
```

```python
for col in shared_ahs.columns:
    try:
        shared_ahs[col] = shared_ahs[col].str.split("'").str[1].tolist()
    except:
        continue
    shared_ahs[col].astype(str)
```

```python
shared_ahs
```

```python
mapped_ahs = shared_ahs
for col in shared_ahs.columns:
    row = relevant_col_df.loc[relevant_col_df['Variable'] == col]
    desc = row['Description'].values[0]
    codes = row['Response Codes'].values[0].split('||')
    map_ = {}
    for code in codes:
        [k, v] = code.split(':', 1)
        if 'or' in k:
            [k1, k2] = k.split(' or ')
            k1 = k1.strip()
            k2 = k2.strip()
            map_[k1] = v
            map_[k2] = v
            continue
        k = k.strip()
        map_[k] = v
    mapped_ahs[col] = shared_ahs[col].map(map_).fillna(shared_ahs[col])
```

```python
mapped_ahs['HINCP'] = mapped_ahs['HINCP'].astype(int)
mapped_ahs['FINCP'] = mapped_ahs['FINCP'].astype(int)
mapped_ahs
```

```python
# Now that data is clean, start analysis.
income_bins = [0, 20000, 40000, 60000, 80000, 100000, 120000, 140000, float('inf')]
regions = ["National",
           " New England",
           " Middle Atlantic",
           " East North Central",
           " West North Central",
           " South Atlantic",
           " East South Central",
           " West South Central",
           " Mountain",
           " Pacific"]
modes_of_analysis = ['HEATTYPE',
                     'HEATFUEL',
                     'HOTWATER',
                     'COOKFUEL',
                     'DRYER']
```

```python
# get percentage breakdowns for a type of fuel source
def percent_df(df, income_bins):
    percent_dict = {}
    for j, i in enumerate(income_bins):
        percent_list = []
        for row in df[i+1]:
            percent_list.append(row / df[i+1].sum())
        if j < len(income_bins) - 1:
            percent_dict[f'income ${i}-{income_bins[j+1]}'] = percent_list
#         else:
#             percent_dict[f'income ${i}+'] = percent_list
    return pd.DataFrame(percent_dict, index=df.index)
```

```python
# Step 1.1: How many households are there in each income bracket?
mapped_ahs.HINCP.groupby(pd.cut(mapped_ahs.HINCP, income_bins)).size()
```

```python
# Step 1.2: How many households are there total?
mapped_ahs.HINCP.groupby(pd.cut(mapped_ahs.HINCP, income_bins)).size().sum()
```

```python
def conduct_analysis(income_bins):
    analysis_dict = {}
    
    for region in regions:
        analysis_dict[region] = {}
        for mode in modes_of_analysis:
            analysis_df = pd.DataFrame()
            if region != "National":
                analysis_df = mapped_ahs.loc[mapped_ahs.DIVISION == region][['HINCP', mode]]
            else:
                analysis_df = mapped_ahs[['HINCP', mode]]
            analysis_groups = analysis_df.groupby([mode, pd.cut(analysis_df.HINCP, income_bins)]).size().unstack()
            analysis_groups_percent = percent_df(analysis_groups, income_bins)
            total_by_income = analysis_groups.sum(axis=1)
            
            analysis_dict[region][mode] = [analysis_groups, analysis_groups_percent, total_by_income]
    return analysis_dict

analysis_dict = conduct_analysis(income_bins)
```

```python
# I've abstracted all of the logic from my analysis into a function and applied it programmatically. 
# Here's how to use this dict:
#   analysis_dict[<region>][<mode>][<0, 1, 2>]
#   regions: see the region list above. leading space is necessary as I haven't been able to get it to stop showing up.
#   mode: see mode list above.
#   <0, 1, 2>:
#        0 — raw counts of each category within each income bracket
#        1 — percentages of each category within each income bracket
#        2 — raw count of each category without income analysis
```

```python
# interpolate AHS numbers of households with percentages of people with each kind of appliance in each region

region_dict = {}
for region in analysis_dict2.keys():
    if region == 'National':
        continue
    region_dict[region] = {}
    for mode in analysis_dict2[region].keys():
        mult = analysis_dict2[region][mode][1].to_numpy() * ahs_household_df.loc[[region]].to_numpy()
        mult = pd.DataFrame(mult,
                            index=analysis_dict2[region][mode][1].index,
                            columns=analysis_dict2[region][mode][1].columns)
        mult['Total'] = mult.sum(axis=1)
        region_dict[region][mode] = mult
```

```python
import csv

for region in region_dict.keys():
    print(region)
    heat_df = region_dict[region]['HEATFUEL']
    heat_df.loc['Electric heat pump'] = region_dict[region]['HEATTYPE'].loc[[" Electric heat pump"]].values[0]
    heat_df.loc['Electric non-heat pump'] = heat_df.loc[[' Electricity']].values[0] - heat_df.loc[['Electric heat pump']].values[0]
    heat_df.loc['Fuel oil/kerosene'] = heat_df.loc[[' Fuel oil']].values[0] + heat_df.loc[[' Kerosene or other liquid fuel']].values[0]
    heat_df.loc['Space heat non-heat pump fraction'] = heat_df.loc[['Electric non-heat pump']].values[0] / heat_df.loc[[' Electricity']].values[0]
    heat_df = heat_df.loc[[' Piped gas',
                           ' LP (liquid petroleum) gas',
                           'Fuel oil/kerosene',
                           ' Electricity',
                           'Electric heat pump',
                           'Electric non-heat pump',
                           'Space heat non-heat pump fraction',
                          ]]
    
    water_df = region_dict[region]['HOTWATER']
    water_df.loc['Electric heat pump'] = water_df.loc[[' Has running hot water with electric water heater']].values[0] * .011
    water_df.loc['Electric non-heat pump'] = water_df.loc[[' Has running hot water with electric water heater']].values[0] * .989
    water_df.loc['Water heat non-heat pump fraction'] = water_df.loc[['Electric non-heat pump']].values[0] / water_df.loc[[' Has running hot water with electric water heater']].values[0]
    try:
        water_df.loc['Fuel oil'] = water_df.loc[[' Has running hot water with fuel oil water heater']].values[0]
    except:
        water_df.loc['Fuel oil'] = [0, 0, 0, 0, 0, 0, 0, 0]
    water_df = water_df.loc[[' Has running hot water with piped gas water heater',
                             ' Has running hot water with LP gas water heater',
                             'Fuel oil',
                             ' Has running hot water with electric water heater',
                             'Electric heat pump',
                             'Electric non-heat pump',
                             'Water heat non-heat pump fraction'
                            ]]
    
    cook_df = region_dict[region]['COOKFUEL'].loc[[' Piped gas',
                                                   ' LP gas (liquid propane)',
                                                   ' Other',
                                                   ' Electric'
                                                  ]]
    
    dryer_df = region_dict[region]['DRYER'].loc[[' Working piped gas powered clothes dryer',
                                                   ' Working LP (liquefied petroleum) gas clothes dryer',
                                                   ' Working electric power clothes dryer'
                                                  ]]
    
    to_csv = ''
    to_csv += heat_df.to_csv()
    to_csv += ',,,,,,,,\n,,,,,,,,\n,,,,,,,,\n'
    to_csv += water_df.to_csv(header=False)
    to_csv += ',,,,,,,,\n,,,,,,,,\n,,,,,,,,\n'
    to_csv += cook_df.to_csv(header=False)
    to_csv += ',,,,,,,,\n,,,,,,,,\n,,,,,,,,\n'
    to_csv += dryer_df.to_csv(header=False)
    
    with open(f'./output/{region[1:]}.csv', "wt") as file:
        file.write(to_csv)
    
```
